```python
def non_regional_FDI(numpy_x, dir_name, region_classif_df, code_map_dict):
    home_code = numpy_x[0]
    host_code = numpy_x[1]
    home_region = numpy_x[2]
    host_region = numpy_x[3]
    yr = numpy_x[4]
    
    if yr < 2001 or yr> 2012:

        return np.nan
    try:
        df = pd.read_excel(dir_FDI_long +home_code +'_long.xls')
    except:
        print('Not found for: ',home_code)
        return np.nan
        
    df.home = df.home.str.strip()
    df['code'] = df['home'].map(country_code_map)
    df.dropna(subset=['code'], inplace = True)
    
    df_fdi = pd.merge(df, region_classif_df, how='left', on=['code'] )
    print(yr)
    print('Hell')
    print(home_code, '-', host_code)
    year_sum = df_fdi[(df_fdi.region !=  home_region) & (df_fdi.code != host_code )].groupby(['year']).fdi.sum()
    print('yay')
    if len(year_sum) == 0:
        return 0.0
    else:
        return year_sum.loc[yr]
    
```
