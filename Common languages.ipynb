```python
lang_dict = {}
for index, row in df_lang.iterrows():
    country_code = row[2]
    lang_text = row[1]
    lang_list = [x.strip() for x in lang_text.split(',')]
    lang_dict[country_code] = lang_list
    
```

```python
lang_dict
```

```python
import math
def common_language(x):
    codeA = x[0]
    codeB = x[1]
    try:
        langA = lang_dict[codeA]
        langB = lang_dict[codeB]
    except:
        return math.nan
    for lang in langA:
        if lang in langB:
            return int(1)
    return int(0)
    
```
