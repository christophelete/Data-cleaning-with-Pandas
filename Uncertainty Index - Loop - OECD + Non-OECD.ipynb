```python
directory  = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/Data/Trends_copy/"
saving_dir = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/Data/ALL_uncertainty/"
```

```python
files = []
for path in os.listdir(directory):
    files.append(path)
len(files)
```

```python
for path in os.listdir(directory):
    country_codes = re.findall(r'([A-Z]+)', path)
    home_3dig = country_codes[0]
    host_3dig = country_codes[1]
    
    try: 
        df = pd.read_csv(directory + path, parse_dates= ['date'])
    except ValueError:
        print('No data for search term of host country:')
        print(home_3dig,"'s interests in",host_3dig)
        print('_______________________________________')
        continue
    if len(df.columns) < 4:
        print('Something is not right for: ')
        print('')
        print(home_3dig+'_'+host_3dig+'.csv')
        break
        
    home_code = df.iloc[0,2]
    host_code = df.iloc[0,3]
    if home_code != home_3dig:
        raise ValueError('Home country code file name does not match home column!')
    if host_code != host_3dig:
        raise ValueError('Host country code file name does not match host column!')
        
    df_index = df.groupby(df.date.dt.year).var()
    search_term = df_index.columns[0]
    df_index.rename(columns = {search_term: 'uncertainty'}, inplace = True)
    df_index['home'] = home_3dig
    df_index['host'] = host_3dig
    df_index.reset_index(inplace = True)
    df_index.to_csv(saving_dir +home_code+'_' + host_code +'_'+'uncertainty.csv', index = False)
```
