```python
def neighbor_dist(x, rank_number):
    sort_x = np.sort(x)
    return sort_x[rank_number]
```

```python
closest_neighbor1 = df_dist_long.groupby(['home']).distance.agg(neighbor_dist,0).to_frame().reset_index()
closest_neighbor2 = df_dist_long.groupby(['home']).distance.agg(neighbor_dist,1).to_frame().reset_index()
closest_neighbor3 = df_dist_long.groupby(['home']).distance.agg(neighbor_dist,2).to_frame().reset_index()
closest_neighbor4 = df_dist_long.groupby(['home']).distance.agg(neighbor_dist,3).to_frame().reset_index()
closest_neighbor5 = df_dist_long.groupby(['home']).distance.agg(neighbor_dist,4).to_frame().reset_index()
```
