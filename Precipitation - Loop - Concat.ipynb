```python
import pandas as pd
import numpy as np
import os
import re
```

```python
directory  = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/Data/"
dir_precip = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/Data/Precipitation/"
```

```python
df_concat = pd.DataFrame()
for path in os.listdir(dir_precip):
    location_name = path.split('.')[-3]
    location_name = location_name.replace('_',' ')
    location_name = location_name.replace('Isl','Island')
    #location_name.replace('Isl')
    print(location_name)
    df = pd.read_csv(dir_precip + path,  delim_whitespace=True ,skiprows=3)
    df_sub = df[df.columns.difference(['MAM','JJA', 'SON', 'DJF','ANN'])]
    
    df_cleaned = pd.DataFrame()
    df_cleaned['year'] = df_sub['YEAR']
    df_cleaned['country'] = location_name
    df_cleaned['monthly_precip'] = df_sub.mean(axis=1)
    df_cleaned['yearly_precip'] = df_sub.mean(axis=1) *12
    df_cleaned['precip_volatility'] = df_sub.var(axis =1)
    df_concat = pd.concat([df_concat,df_cleaned])
    
```

```python
df_concat.reset_index(drop=True, inplace= True)
```
