```python
FDI_long_dir = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/Data/Long_FDI/"
directory = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/"
data_dir  = "/media/sony/0CA83E99A83E80EE/Chris/Bocconi/Thesis/Data_copy/Data/FDI"
```

```python
#generate country code map
names_convertAB = {'Congo, Democratic Rep. of' : 'Democratic Republic of Congo',
           'Venezuela, Bolivarian Rep. of': 'Bolivarian Republic of Venezuela',
          'United Rep. of Tanzania': 'United Republic of Tanzania',
           'Korea, Republic of': 'Republic of Korea',
           'Iran, Islamic Republic of': 'Islamic Republic of Iran',
           'Moldova, Republic of': 'Republic of Moldova',
           'Bolivia, Plurinational State of': 'Plurinational State of Bolivia',
            "Lao People's Dem. Rep.": "Lao People's Democratic Republic"
          }
names_convertBA = dict( [(value, key) for key, value in names_convertAB.items()] )
country_code_map = dict()
for path in os.listdir(data_dir):
    country_code = re.findall(r'_([A-Z]+)', path)[0]
    hostname_df = pd.read_excel(data_dir+'/' + path, sheet_name= 0, na_values ='..')
    country_code_map[" ".join(hostname_df.columns[0].split())] = country_code
for key, val in names_convertBA.items():
    country_code_map[val] = country_code_map[key]
    del country_code_map[key]
```

```python
df_base = pd.read_csv(directory +'base_dataset.csv')
df_base_copy = df_base.copy()
```

```python
count = 0
for path in os.listdir(FDI_long_dir):
    print(path)
    country_code = re.findall(r'([A-Z]+)', path)[0]
    print(country_code)
    df_merging = pd.read_excel(FDI_long_dir + path)
    length_merging = len(df_merging[df_merging.home.isin(country_code_map.keys())]) #count only countries, not continents
    listed_countries = df_merging.home
    if count ==0:
        df_merged = pd.merge(df_base_copy,df_merging,how = 'left', on = ['home','host','year'])
    else:
        df_merged = pd.merge(df_merged,df_merging,how = 'left', on = ['home','host','year'])
        df_merged['fdi'] = np.where(df_merged['fdi_x'].isna(),df_merged['fdi_y'],df_merged['fdi_x'])
        del df_merged['fdi_x']
        del df_merged['fdi_y']
        
    count += 1
```
