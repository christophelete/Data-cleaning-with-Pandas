```python
df_main['union_partner'] = df_main[['EU_partner','asean_partner',
                                    'mercosour_partner']].apply(lambda x: int(1)if x[0]== 1 or x[1]==1 or x[2]==1 
                                                                else int(0), axis =1)
```

```python
def isNAFTA(x):
    if x[0] in ['MEX','USA', 'CAN'] and x[1] in ['MEX','USA', 'CAN'] and x[2] >= 1994:
        return int(1)
    else:
        return int(0)
```

```python
df_main['NAFTA'] = df_main[['home_code','host_code','year']].apply(isNAFTA,axis = 1)
```

```python
df_mig_sub = df_migration[['Country Origin Code','Country Dest Code','1993','2000','2001','2002','2003']]
```

```python
df_merged = pd.merge(df_main,df_mig_melt, how='left', on =['home_code', 'host_code', 'year'])
```
