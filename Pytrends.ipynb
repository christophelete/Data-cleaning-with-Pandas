```python
already_scraped = dict()
for path in os.listdir(directory):
    country_codes = re.findall(r'([A-Z]+)', path)
    try:
        already_scraped[country_codes[0]].append(country_codes[1])
    except:
        already_scraped[country_codes[0]] = [country_codes[1]]
        
    
```
